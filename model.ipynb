```python
x = df[['views','dislikes','comment_count','Genre']]
```

```python
x
```

```python
y = df['likes']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

var = lr.score(x_test, y_test)
var
```

```python
bias = lr.score(x_train, y_train)
bias
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
import pickle
with open('model.pickle', 'wb') as model_file:           # wb is for we are storing the data in binary code
    pickle.dump(lr, model_file)           # dumping the model to the file
```
